```python
import numpy as np
from matplotlib import pyplot as plt
from fractions import Fraction 
x=np.arange(0, 1.01, 0.01)
plt.figure(dpi=600)
c=0.5
y1 = []
y2 = []
for p in x:
    y_1 = 1/np.sqrt(2)*np.sqrt((np.sqrt(p)-np.sqrt(c))**2+(np.sqrt(1-p)-np.sqrt(c))**2)
    y1.append(y_1)
    y_2 = 1/np.sqrt(2)*np.sqrt((np.sqrt(0.999)-np.sqrt(c))**2+(np.sqrt(0.001)-np.sqrt(c))**2)
    y2.append(y_2)

plt.scatter(x, y1,s=6,  marker = '*',label='$\it{H}$($\it{m}$$_1$)')
plt.scatter(x, y2,s=6,  marker = '<',label='$\it{H}$($\it{m}$$_2$)')
plt.xlabel('a',fontproperties="Times New Roman",fontstyle='italic',fontsize=12,fontweight='bold')
plt.ylabel('H',fontproperties="Times New Roman",fontstyle='italic',fontsize=12,fontweight='bold')

plt.yticks(fontproperties='Times New Roman', size=12)#
plt.xticks(fontproperties='Times New Roman', size=12)
font1 = {'family':'Times New Roman',
'weight':'normal',
'size': 12,
}


plt.legend(handletextpad=0.1,prop=font1)   #
plt.savefig("Fig.2.pdf")
plt.show()
```
